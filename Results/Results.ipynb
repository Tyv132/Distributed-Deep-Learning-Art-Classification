```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 900
plt.rcParams['savefig.dpi'] = 900
```

```python
data = pd.read_csv('gpus.csv')
```

```python
sns.pointplot(x='n_gpus', y='msec_per_img', hue='batch_size_global',
    data=data, dodge=True, join=False, markers='.', palette=['#f0f921', '#fca636', '#e16462'])
plt.legend(title='Batch size (global)', loc=(1.02, 0.52), frameon=False)
plt.xlabel('Number of GPUs')
plt.ylabel('Training time per image (ms)');
plt.ylim(0, 30)
```

```python
data = data[['n_gpus', 'batch_size_global', 'val_accuracy']]
```

```python
plt.plot([2, 3, 4, 5, 6, 7], data[0:6].val_accuracy, 'o-', label='1 GPU: 32 images', color='#f0f921')
plt.plot([2, 3, 4, 5, 6, 7], data[6:12].val_accuracy, 'o-', label='1 GPU: 64 images', color='#fca636')
plt.plot([2, 3, 4, 5, 6, 7], data[12:18].val_accuracy, 'o-', label='1 GPU: 128 images', color='#e16462')
plt.plot([2, 3, 4, 5, 6, 7], data[18:24].val_accuracy, 'o-', label='2 GPU: 32 images', color='#b12a90')
plt.plot([2, 3, 4, 5, 6, 7], data[24:30].val_accuracy, 'o-', label='2 GPU: 64 images', color='#6a00a8')
plt.plot([2, 3, 4, 5, 6, 7], data[30:36].val_accuracy, 'o-', label='2 GPU: 128 images', color='#0d0887')
plt.legend(title='Batch size (global)', loc=(1.02, 0.35), frameon=False)
plt.xlabel('Epoch')
plt.ylabel('Validation accuracy (%)');
plt.ylim(0, 50)
```
